```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
from scipy.interpolate import UnivariateSpline
import os
from google.colab import drive
import tensorflow as tf
from tensorflow import keras
from sklearn.preprocessing import StandardScaler

# Biological properties of tissues
TISSUE_PROPERTIES = {
    'esophagus.csv': {
        'name': 'Mouse Esophagus',
        'type': 'GI tract organ',
        'structure': 'stratified squamous epithelium',
        'anisotropy': 'moderate',
        'fiber_direction': 'circular'
    },
    'skin.csv': {
        'name': 'Mouse Skin',
        'type': 'cover tissue',
        'structure': 'stratified epithelium + connective tissue',
        'anisotropy': 'weak',
        'fiber_direction': 'random'
    },
    'intensive_large-1.csv': {
        'name': 'Mouse Large Intestine',
        'type': 'GI tract organ',
        'structure': 'simple columnar epithelium',
        'anisotropy': 'high',
        'fiber_direction': 'longitudinal and circular'
    },
    'intensive_thin1.csv': {
        'name': 'Mouse Small Intestine (Sample 1)',
        'type': 'GI tract organ',
        'structure': 'simple columnar epithelium with villi',
        'anisotropy': 'high',
        'fiber_direction': 'longitudinal'
    },
    'intensive_thin2.csv': {
        'name': 'Mouse Small Intestine (Sample 2)',
        'type': 'GI tract organ',
        'structure': 'simple columnar epithelium with villi',
        'anisotropy': 'high',
        'fiber_direction': 'longitudinal'
    }
}

def mount_google_drive():
    """Connect to Google Drive"""
    drive.mount('/content/drive')

def save_to_google_drive(data, filename, folder='/content/drive/MyDrive/Biomechanics_Data'):
    """Save data to Google Drive"""
    os.makedirs(folder, exist_ok=True)
    full_path = os.path.join(folder, filename)
    data.to_csv(full_path, index=False)
    print(f"Data saved to: {full_path}")

def load_and_preprocess_data(file_path):
    """Load and preprocess data"""
    with open(file_path, 'r', encoding='utf-8-sig') as f:
        content = f.read().replace(',', '.')

    data = pd.read_csv(pd.io.common.StringIO(content), delimiter=';')
    data = data[['Strain', 'Stress']].astype(float)
    data = data.drop_duplicates('Strain', keep='first').sort_values('Strain')

    return data

def smooth_experimental_data(data, method='spline', **kwargs):
    strain = data['Strain'].values
    stress = data['Stress'].values
    spl = UnivariateSpline(strain, stress, **kwargs)
    return pd.DataFrame({'Strain': strain, 'Stress': spl(strain)})

def generate_synthetic_data(original_data, num_samples=19, noise_level=0.03):
    smoothed_data = smooth_experimental_data(original_data, s=0.5)
    synthetic_data = []

    for _ in range(num_samples):
        synth_sample = smoothed_data.copy()
        noise = np.random.normal(0, noise_level*smoothed_data['Stress'].std(), len(smoothed_data))
        synth_sample['Stress'] = smoothed_data['Stress'] * (1 + noise)
        synthetic_data.append(synth_sample)

    return [original_data] + synthetic_data

def calculate_average_curve(data_list):
    min_strain = max([d['Strain'].min() for d in data_list])
    max_strain = min([d['Strain'].max() for d in data_list])
    common_strain = np.linspace(min_strain, max_strain, 100)

    interp_stresses = []
    for data in data_list:
        interp_stress = np.interp(common_strain, data['Strain'], data['Stress'])
        interp_stresses.append(interp_stress)

    avg_stress = np.mean(interp_stresses, axis=0)
    return common_strain, avg_stress

# Hyperelastic models considering biological properties

def neo_hookean(strain, C1):
    lmbda = 1 + strain/100
    return 2*C1*(lmbda - 1/lmbda**2)

def mooney_rivlin_2param(strain, C1, C2):
    lmbda = 1 + strain/100
    return 2*(lmbda - 1/lmbda**2)*(C1 + C2/lmbda)

def mooney_rivlin_3param(strain, C1, C2, C3):
    lmbda = 1 + strain/100
    return 2*(lmbda - 1/lmbda**2)*(C1 + 2*C2*(lmbda**2 + 2/lmbda - 3) + C3/lmbda)

def yeoh_2order(strain, C1, C2):
    lmbda = 1 + strain/100
    I1 = lmbda**2 + 2/lmbda
    return 2*(lmbda - 1/lmbda**2)*(C1 + 2*C2*(I1-3))

def yeoh_3order(strain, C1, C2, C3):
    lmbda = 1 + strain/100
    I1 = lmbda**2 + 2/lmbda
    return 2*(lmbda - 1/lmbda**2)*(C1 + 2*C2*(I1-3) + 3*C3*(I1-3)**2)

def ogden_isotropic(strain, mu1, alpha1, mu2=None, alpha2=None):
    lmbda = 1 + strain/100
    if mu2 is None or alpha2 is None:
        return mu1*(lmbda**(alpha1-1) - lmbda**(-alpha1/2-1))
    else:
        return (mu1*(lmbda**(alpha1-1) - lmbda**(-alpha1/2-1)) +
                mu2*(lmbda**(alpha2-1) - lmbda**(-alpha2/2-1)))

def ogden_anisotropic(strain, mu, alpha, k1, k2, fiber_direction='circular'):
    lmbda = 1 + strain/100

    # Basic isotropic contribution
    iso_term = mu*(lmbda**(alpha-1) - lmbda**(-alpha/2-1))

    # Anisotropic contribution (depends on fiber direction)
    if fiber_direction == 'circular':
        aniso_term = k1*(lmbda**(2*k2) - lmbda**(-k2))
    else:
        aniso_term = k1*(lmbda**(3*k2) - lmbda**(-0.5*k2))

    return iso_term + aniso_term

def hgo_model(strain, k1, k2, kappa=0.05):
    lmbda = 1 + strain/100
    I1 = lmbda**2 + 2/lmbda
    I4 = kappa*I1 + (1-3*kappa)*lmbda**2
    return 2*(lmbda - 1/lmbda**2)*(1 + 2*k1*(I4-1)*np.exp(k2*(I4-1)**2)*(1-3*kappa))

# Model fitting

def fit_and_plot_models(strain, stress, tissue_props):
    models = []

    # Basic models (suitable for all tissues)
    models.append(('Neo-Hookean', neo_hookean, [1.0]))
    models.append(('Mooney-Rivlin (2p)', mooney_rivlin_2param, [1.0, 0.1]))

    # Models for tissues with pronounced anisotropy
    if tissue_props['anisotropy'] in ['moderate', 'high']:
        models.append(('Ogden Anisotropic',
                     lambda x, a, b, c: ogden_anisotropic(x, a, b, c, 0.1, tissue_props['fiber_direction']),
                     [1.0, 2.0, 0.5]))
        models.append(('HGO Model', hgo_model, [0.1, 0.1]))

    # For GI tract use more complex models
    if 'GI tract' in tissue_props['type']:
        models.append(('Mooney-Rivlin (3p)', mooney_rivlin_3param, [1.0, 0.1, 0.01]))
        models.append(('Yeoh (3rd order)', yeoh_3order, [1.0, 0.1, 0.01]))

    # For skin use simpler models
    if 'skin' in tissue_props['name'].lower():
        models.append(('Ogden Isotropic (2p)', lambda x, a, b: ogden_isotropic(x, a, b), [1.0, 2.0]))

    plt.figure(figsize=(14, 8))
    plt.plot(strain, stress, 'ko', markersize=6, label='Averaged data')

    results = {}
    for name, model, init_params in models:
        try:
            params, _ = curve_fit(model, strain, stress, p0=init_params, maxfev=5000)
            fitted = model(strain, *params)

            plt.plot(strain, fitted, linewidth=2, label=f'{name} fit')

            results[name] = {
                'params': params,
                'fitted': fitted,
                'error': np.mean(np.abs(fitted - stress))
            }

            print(f"{name}: params={params}, error={results[name]['error']:.4f}")
        except Exception as e:
            print(f"Error fitting {name}: {str(e)}")

    plt.xlabel('Strain (%)', fontsize=12)
    plt.ylabel('Stress (MPa)', fontsize=12)
    plt.title(f'Hyperelastic model fitting for {tissue_props["name"]}\n'
              f'Structure: {tissue_props["structure"]}, Anisotropy: {tissue_props["anisotropy"]}',
              fontsize=14)
    plt.legend(fontsize=10, bbox_to_anchor=(1.05, 1), loc='upper left')
    plt.grid(True)
    plt.tight_layout()
    plt.show()

    return results

# Neural network modeling

def create_pinn_model():
    model = keras.Sequential([
        keras.layers.Dense(128, activation='tanh', input_shape=(1,)),
        keras.layers.Dense(128, activation='tanh'),
        keras.layers.Dense(128, activation='tanh'),
        keras.layers.Dense(1)
    ])
    model.compile(optimizer='adam', loss='mse')
    return model

def train_and_predict_pinn(strain, stress, tissue_name):
    X = strain.reshape(-1, 1)
    y = stress

    scaler = StandardScaler()
    X_scaled = scaler.fit_transform(X)
    y_scaled = (y - y.mean()) / y.std()

    model = create_pinn_model()
    history = model.fit(X_scaled, y_scaled, epochs=1000, verbose=0, validation_split=0.2)

    strain_range = np.linspace(strain.min(), strain.max()*1.1, 100)
    predicted = model.predict(scaler.transform(strain_range.reshape(-1, 1)).flatten())
    predicted = predicted * y.std() + y.mean()

    plt.figure(figsize=(12, 6))
    plt.plot(strain, stress, 'ko', label='Experimental data')
    plt.plot(strain_range, predicted, 'r-', linewidth=2, label='PINN prediction')
    plt.xlabel('Strain (%)')
    plt.ylabel('Stress (MPa)')
    plt.title(f'Neural network (PINN) prediction for {tissue_name}')
    plt.legend()
    plt.grid(True)
    plt.show()

    return model, strain_range, predicted

def process_material(file_path):
    """Full processing pipeline for one material"""
    tissue_props = TISSUE_PROPERTIES[os.path.basename(file_path)]
    print(f"\n{'='*50}")
    print(f"Processing tissue: {tissue_props['name']}")
    print(f"Type: {tissue_props['type']}")
    print(f"Structure: {tissue_props['structure']}")
    print(f"Anisotropy: {tissue_props['anisotropy']}")
    print(f"Fiber direction: {tissue_props['fiber_direction']}")
    print(f"{'='*50}")
    print("\n1. Loading data...")
    try:
        original_data = load_and_preprocess_data(file_path)
        print("Sample data:")
        print(original_data.head())
    except Exception as e:
        print(f"Loading error: {str(e)}")
        return None
    print("\n2. Generating synthetic data...")
    all_data = generate_synthetic_data(original_data)
    print("\n3. Calculating average curve...")
    avg_strain, avg_stress = calculate_average_curve(all_data)
    avg_data = pd.DataFrame({'Strain': avg_strain, 'Stress': avg_stress})
    print("\n4. Fitting hyperelastic models...")
    model_results = fit_and_plot_models(avg_strain, avg_stress, tissue_props)
    print("\n5. Neural network modeling...")
    pinn_model, pinn_strain, pinn_stress = train_and_predict_pinn(avg_strain, avg_stress, tissue_props['name'])
    print("\n6. Saving results...")
    save_to_google_drive(avg_data, f"{tissue_props['name']}_average.csv")

    return {
        'tissue_properties': tissue_props,
        'average_data': avg_data,
        'model_results': model_results,
        'pinn_prediction': (pinn_strain, pinn_stress)
    }

def main():
    """Main function to process all materials"""
    print("Connecting to Google Drive...")
    mount_google_drive()
    results = {}
    for file_name, props in TISSUE_PROPERTIES.items():
        try:
            results[props['name']] = process_material(file_name)
        except Exception as e:
            print(f"\nError processing {props['name']}: {str(e)}")

    print("\nAll tissues processed successfully!")
    return results

if __name__ == "__main__":
    all_results = main()
```
